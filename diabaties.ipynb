```python
#importing all important module for data manipulation
import numpy as np 
import pandas as pd 
```

```python
#importing all important module for model training
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#access dataset by using pandas 
df = pd.read_csv("diabetes.csv")
```

```python
#printing the dataset we accessing with the help of pandas 
df.head()
```

```python
# to get rows and column of our dataset 
df.shape
```

```python
#getting the statiscal measure of the dataset 
df.describe()
```

```python
# checking if there was a missing value 
df.isna()
```

```python
df['Outcome'].value_counts()
```

```python
#separating the data and the labels
x= df.drop(columns='Outcome', axis=1)
y = df['Outcome']
```

```python
#standardrisation 
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standard_data = scaler.fit_transform(x)
```

```python
x = standard_data
```

```python
x
```

```python
# train and test 
x_train , x_test , y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#train the model 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = LogisticRegression()
```

```python
rf_model = RandomForestClassifier(random_state=42)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pre = clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pre)
```

```python
import numpy as np
import pickle

# Load the model and scaler
with open("trained_model_with_scaler.sav", "rb") as f:
    loaded_model, scaler = pickle.load(f)

# Input data for prediction
input_data = (0, 131, 0, 0, 0, 43.2, 0.27, 26)

# Convert input data to numpy array and reshape it
input_data_reshaped = np.asarray(input_data).reshape(1, -1)

# Standardize the input data using the loaded scaler
std_data = scaler.transform(input_data_reshaped)

# Predict using the loaded model
prediction = loaded_model.predict(std_data)

if prediction[0] == 0:
    print("This is not a diabetic person")
else:
    print("This person is diabetic")
```
